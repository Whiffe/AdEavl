```python
import matplotlib.pyplot as plt
import numpy as np

# 设置全局字体大小为10pt
plt.rcParams.update({'font.size': 14})

# Data
x = [1, 2, 3, 4, 5]
org = [80.7, 71.6, 65.4, 65.8, 60.4]
DC = [81, 74.6, 70.8, 66.2, 64.6]
threshold = 60.9

# Plotting
plt.figure(figsize=(6, 4))
plt.plot(x, org, marker='o', linestyle='--', color='b', label='CLN')
plt.plot(x, DC, marker='s', linestyle='-', color='g', label='CTM')
plt.axhline(y=threshold, color='r', linestyle='-.', label='Threshold = 60.9')

# Labels and Title
plt.xlabel('Reconstruction times')
plt.ylabel('Precision')
# plt.title('Multiple rounds of reconstruction adjust the problem complexity')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

# 设置横坐标轴刻度格式为整数
plt.xticks(x)

# 保存为PDF文件
plt.savefig('Reconstruction_Complexity.pdf', format='pdf', bbox_inches='tight')  # 保存为PDF，确保内容完整

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams

# 设置字体为Times New Roman，字体大小为11pt
rcParams['font.family'] = 'Times New Roman'
rcParams['font.size'] = 8

# 第一组数据
datasets1 = ['Static', 'AdEval', 'Auto-Dataset', 'LatestEval']
models1 = ['Qwen2-7B-Chat', 'Llama3-8B-Instruct', 'glm-4-9b-chat']
cln_data1 = [
    [60.9, 60.4, 76.5, 84.2],
    [55.5, 60.8, 66.0, 79.2],
    [64.7, 66.5, 66.0, 77.5]
]
ctm_data1 = [
    [85.2, 64.6, 68.4, 86.6],
    [84.4, 63.6, 71.0, 79.6],
    [90.3, 62.9, 73.2, 81.3]
]
errors_cln1 = [
    [0.8, 0.6, 0.8, 0.4],
    [1.2, 1.0, 1.0, 0.4],
    [1.0, 1.0, 2.0, 0.3]
]
errors_ctm1 = [
    [1.9, 1.6, 0.7, 0.8],
    [3.4, 0.7, 1.7, 0.4],
    [1.0, 1.1, 0.9, 0.5]
]

# 第二组数据
datasets2 = ['Static', 'AdEval', 'Auto-Dataset', 'LatestEval']
models2 = ['Qwen2-7B-Chat', 'Meta-Llama3-8B-Instruct', 'glm-4-9b-chat']
cln_data2 = [
    [83.0, 81.4, 86.4, 88.4],
    [73.1, 74.6, 82.0, 86.0],
    [83.8, 82.7, 89.7, 90.7]
]
ctm_data2 = [
    [99.0, 81.1, 90.5, 92.6],
    [96.3, 73.8, 88.8, 88.7],
    [94.6, 82.7, 89.5, 87.8]
]
errors_cln2 = [
    [1.2, 0.8, 0.7, 0.4],
    [0.9, 0.5, 0.0, 0.4],
    [1.3, 1.3, 1.0, 0.4]
]
errors_ctm2 = [
    [0.3, 1.3, 0.8, 0.2],
    [0.4, 0.7, 0.5, 0.7],
    [0.4, 0.3, 1.5, 0.3]
]

# 合并数据列表
all_cln_data = [cln_data1, cln_data2]
all_ctm_data = [ctm_data1, ctm_data2]
all_errors_cln = [errors_cln1, errors_cln2]
all_errors_ctm = [errors_ctm1, errors_ctm2]

# 设置柱状图的宽度
bar_width = 0.35  # 设置柱状图的宽度

# 设置每个柱状图的位置
r1 = np.arange(len(datasets1))
r2 = [x + bar_width for x in r1]

# 创建子图，2行3列
fig, axs = plt.subplots(2, 3, figsize=(7, 3), sharey=False)

# 颜色和样式
colors = ['#0000FF', '#4692f5', '#ff7f0e', '#ffbb78', '#2ca02c', '#98df8a']
patterns = ['//', '', '//', '', '//', '']

# 在第一行和第二行左边添加竖着的标签，调整x坐标
fig.text(0.01, 0.75, 'MMLU', rotation=90, va='center', ha='center', fontsize=8)
fig.text(0.01, 0.25, 'ARC-Challenge', rotation=90, va='center', ha='center', fontsize=8)

# 遍历两组数据
for row in range(2):
    # 遍历每个模型
    for col, model in enumerate(models1 if row == 0 else models2):
        axs[row, col].bar(r1, all_cln_data[row][col], color=colors[col*2], width=bar_width, edgecolor='white', yerr=all_errors_cln[row][col], capsize=3, label=f'CLN', hatch=patterns[col*2])
        axs[row, col].bar(r2, all_ctm_data[row][col], color=colors[col*2+1], width=bar_width, edgecolor='white', yerr=all_errors_ctm[row][col], capsize=3, label=f'CTM', hatch=patterns[col*2+1])
        
        # 设置横坐标刻度标签
        axs[row, col].set_xticks([r + bar_width/2 for r in range(len(datasets1))])
        if row == 0:
            axs[row, col].set_xticklabels([''] * len(datasets1))  # 第一行不显示横坐标内容
        else:
            axs[row, col].set_xticklabels(datasets1, rotation=20, ha="right")  # 斜着显示横坐标的标签
        
        # 根据行号设置不同的纵坐标起点
        if row == 0:
            axs[row, col].set_ylim(45, 100)
        else:
            axs[row, col].set_ylim(65, 100)
        if row == 0:
            axs[row, col].set_title(model)
            axs[row, col].legend()
        # 确保每个子图都设置了纵坐标刻度标签
        axs[row, col].tick_params(axis='y', which='both', labelleft=True)

        # 若为第二列或第三列子图，不显示纵坐标值
        if col > 0:
            axs[row, col].set_yticklabels([])

# 调整子图之间的宽度和高度空间
plt.subplots_adjust(wspace=0.2, hspace=0.1, left=0.07)

# 保存图形为PDF
plt.savefig('Data_Contamination_Fine-Tuning_combined.pdf', bbox_inches='tight')

plt.tight_layout()
plt.show()
```
