```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams

rcParams['font.size'] = 10

#  - 
models = ['gpt-4o-mini', 'DeepSeek-V3', 'Doubao-pro', 'qwen-plus', 'claude-3-5', 'glm-4-flash']
#  - AdEvalAuto - DatasetLatestEval
cln_data = [
    [74.8, 81.0, 87.3],# AdEvalAuto - DatasetLatestEvalA
    [78.0, 82.7, 91.8],
    [84.0, 85.0, 88.9],
    [79.4, 84.4, 90.0],
    [64.8, 75.2, 84.1],
    [66.1, 69.0, 78.5]
]
ctm_data = [
    [74.4, 82.9, 91.4],
    [78.2, 88.0, 94.8],
    [84.6, 87.4, 94.6],
    [79.9, 90.0, 96.1],
    [67.7, 76.0, 88.0],
    [71.4, 72.8, 86.5]
]
# 
errors_cln = [
    [0.0, 0.7, 0.5],
    [0.0, 0.0, 0.0],
    [0.3, 0.7, 0.0],
    [0.2, 0.4, 0.2],
    [0.6, 1.8, 1.6],
    [0.3, 1.3, 0.8]
]
errors_ctm = [
    [0.6, 0.9, 0.4],
    [0.0, 0.0, 0.0],
    [0.3, 0.4, 0.1],
    [0.7, 0.7, 0.3],
    [1.0, 2.0, 0.5],
    [0.4, 1.8, 0.1]
]

datasets = ['EdEval', 'Auto-Dataset', 'LatestEval']
bar_width = 0.35
r1 = np.arange(len(datasets))
r2 = [x + bar_width for x in r1]

# 
num_models = len(models)
num_rows = (num_models + 1) // 2

# num_rows2
# fig, axs = plt.subplots(num_rows, 2, figsize=(5, 2 * num_rows), sharey=True)
fig, axs = plt.subplots(num_rows, 2, figsize=(4, 2 * num_rows))

# axs
if num_rows == 1:
    axs = axs.reshape(1, 2)

colors = ['#0000FF', '#4692f5', '#ff7f0e', '#ffbb78', '#2ca02c', '#98df8a',
          '#d62728', '#ff9896', '#9467bd', '#c5b0d5', '#8c564b', '#c49c94',
          '#228B22', '#FFB6C1']
patterns = ['//', '', '//', '', '//', '', '//', '', '//', '', '//', '', '//', '']

# 
for i, model in enumerate(models):
    row = i // 2
    col = i % 2
    axs[row, col].bar(r1, cln_data[i], color=colors[i*2], width=bar_width, edgecolor='white', yerr=errors_cln[i], capsize=2, label=f'CLN', hatch=patterns[i*2])
    axs[row, col].bar(r2, ctm_data[i], color=colors[i*2+1], width=bar_width, edgecolor='white', yerr=errors_ctm[i], capsize=2, label=f'CTM', hatch=patterns[i*2+1])

    axs[row, col].set_xticks([r + bar_width for r in range(len(datasets))])
    if row == num_rows - 1:
        axs[row, col].set_xticklabels(datasets, rotation=20, ha="right")
    else:
        axs[row, col].set_xticklabels([])
    

    #  70  100
    if row == 0 and col == 0:
        axs[row, col].set_ylim(65, 100)
    if row == 0 and col == 1:
        axs[row, col].set_ylim(70, 100)
    if row == 1:
        axs[row, col].set_ylim(75, 100)
    if row == 2:
        axs[row, col].set_ylim(60, 100)
    if row == 3:
        axs[row, col].set_ylim(60, 100)

    axs[row, col].set_title(model)
    axs[row, col].legend()

# 
if num_models % 2 == 1:
    axs[-1, -1].axis('off')

plt.subplots_adjust(wspace=0.25, hspace=0.2)  # wspace hspace 
plt.savefig('LLMs_test.pdf', bbox_inches='tight')
plt.tight_layout()
plt.show()
```
